```python
#노선 id를 이용하여 데이터 뽑기
def busRouteInfo(busId):
    url = 'http://apis.data.go.kr/6410000/busrouteservice/getBusRouteInfoItem'
    queryParams = '?' + urlencode({ quote_plus('serviceKey') : serviceKey, quote_plus('routeId') : busid })

    request = Request(url + queryParams)
    request.get_method = lambda: 'GET'
    response_body = json.dumps(xmltodict.parse(urlopen(request).read()), ensure_ascii=False)
    return response_body
```

```python
def busRouteList(busNum):
    url = 'http://apis.data.go.kr/6410000/busrouteservice/getBusRouteList'
    queryParams = '?' + urlencode({ quote_plus('serviceKey') : serviceKey, quote_plus('keyword') : busNum })
    request = Request(url + queryParams)
    request.get_method = lambda: 'GET'
    response_body = xmltodict.parse(urlopen(request).read().decode('utf8'))
    return response_body
```
